```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['science', 'notebook','grid'])
```

## 1. Diamonds and weight

```python
def f_linear(weight):
    return -260+3721*weight
x=np.linspace(0, 0.5, 5)
y=f_linear(x)
plt.plot(x, y)
plt.xlabel('Carats')
plt.ylabel('Expected Price')
plt.title('Diamond prices by weight')
plt.show()
```

## 2. Spread of an epidemic

```python
def f_exponential(weeks):
    return 6.69*np.exp(0.18*weeks)
x=np.linspace(0, 45, 100)
y=f_exponential(x)
plt.plot(x, y, 'r')
plt.xlabel('Week')
plt.ylabel('Cases')
plt.title('Ebola Cases')
plt.show()
```

## 3. Demand Models

```python
def f_power(price):
    return 60000*price**(-2.5)
x=np.linspace(1, 1.5, 100)
y=f_power(x)
plt.plot(x, y, 'g')
plt.xlabel('Price')
plt.ylabel('Quantitiy')
plt.title('Demand Model')
plt.show()
```

## 4. The uptake of a new product

```python
def f_prop(year):
    return ( np.e**(2*(year-2.5)) )/ ( 1+np.e**(2*(year-2.5)) )
x=np.linspace(0, 5, 100)
y=f_prop(x)
plt.plot(x, y, 'magenta')
plt.xlabel('Year')
plt.ylabel('Proportion of target population with product')
plt.title('Market Saturation Model')
plt.show()
```
